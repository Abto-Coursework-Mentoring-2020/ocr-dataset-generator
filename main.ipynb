```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
from image_processing import noisify, NoiseTypes

example_image_fp = 'sample_images/text_image.png'
# read image, convert it from BGR to RGB format and create a tensorflow tensor
image = tf.constant(cv2.cvtColor(cv2.imread(example_image_fp), cv2.COLOR_BGR2GRAY))
# plt.imshow(image);

# noise_type is one of SALT_AND_PEPPER, GAUSSIAN, POISSON, SPECKLE
noised = noisify(image, noise_type=NoiseTypes.SALT_AND_PEPPER, salt_vs_pepper=0.1, amount=0.2)
plt.imshow(noised, cmap='gray')
plt.title('Noised image');
```
